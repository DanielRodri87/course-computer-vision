```python
import os
import numpy as np

file = os.path.join('dataset.csv')
```

```python
# Carregando um dataset para dentro de uma lista:
# Observação, só posso pegar elementos do mesmo tipo ao usar o numpy
# Usecols = use colunas 0,1,2,3 porque são do mesmo tipo, a quarta não pode
# skiprows = para começar da segunda linha, porque é string!!
database = np.loadtxt(file, delimiter=',', usecols=(0,1,2,3), skiprows=1)

print(database)
```

```python
# Acima, descarreguei em uma lista, aqui eu descarrego em duas
var1, var2 = np.loadtxt(file, delimiter=',',usecols=(0,1), skiprows=1, unpack=True)
```

```python
# Vamos plotar para visualizar o que acontece
import matplotlib.pyplot as plt
plt.show(plt.plot(var1, var2, 'o', markersize = 6, color = 'red'))
```
